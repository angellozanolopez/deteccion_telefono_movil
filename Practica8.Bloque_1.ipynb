<p>Una vez que hayas accedido a Colab, podrás realizar diversas acciones, incluyendo:</p>
<ul>
  <li><strong>Cargar un vídeo MP4 (o similar)</strong> en el directorio de trabajo de Colab.</li>
  <li><strong>Separar el vídeo en frames</strong> para trabajar con sus componentes individuales.</li>
  <li><strong>Imprimir el número de frames</strong> por el <code>stdout</code> para obtener información sobre la duración y estructura del vídeo.</li>
</ul>


```python
# Importar librerias necesarias
import os
import cv2
import shutil

# Cargar el vídeo desde el directorio local a google colab
video_path = '/content/smartphone.mp4'
cap = cv2.VideoCapture(video_path)

# En caso de que no se pueda cargar el video por algun motivo...
if not cap.isOpened():
    print("Error al abrir el vídeo")
    exit()

# Inicializar el contador de fotogramas
frame_count = 0
# Carpeta de salida donde se almacenaran los fotogramas
output_folder = 'frames'

# Crear la carpeta 'frames' en caso de que no exista aun
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Iterar sobre todos los fotogramas del video
while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    # Incrementar el contador de fotogramas
    frame_count += 1

    # Guardar cada fotograma en la carpeta 'frames'
    cv2.imwrite(os.path.join(output_folder, f"frame_{frame_count}.png"), frame)

# Liberar recursos
cap.release()

# Imprimir el número total de fotogramas
print(f"Número de frames: {frame_count}")
```

<h4>Comprimir carpeta</h4>
<p>Comprimir la carpeta 'fotogramas' para facilitar su descarga al disco local</p>

```python
carpeta_a_comprimir = '/content/frames'

# Ruta donde se guardará el archivo ZIP comprimido
archivo_zip = '/content/frames.zip'

# Comprimir la carpeta 'frames' en un archivo ZIP
shutil.make_archive('/content/frames', 'zip', carpeta_a_comprimir)

# Mover el archivo ZIP a la ubicación deseada
shutil.move('/content/frames.zip', archivo_zip)

# Descargar el archivo ZIP
from google.colab import files
files.download(archivo_zip)
```
